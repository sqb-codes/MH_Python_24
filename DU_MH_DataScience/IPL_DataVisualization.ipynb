```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches_df = pd.read_csv('datasets/ipldata_19/matches.csv')
deliveries_df = pd.read_csv('datasets/ipldata_19/deliveries.csv')
```

```python
matches_df.shape
```

```python
matches_df.isna().sum()
```

```python
pd.value_counts(matches_df['winner'])
```

```python
index = pd.value_counts(matches_df['winner']).index
values = pd.value_counts(matches_df['winner']).values
```

```python
index
```

```python
values
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.bar(index, values, color='red')
plt.title("Winner Count")
plt.xlabel("Team Name")
plt.ylabel("Winning Count")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x = matches_df['winner'])
plt.title("Winner Count")
plt.xlabel("Team Name")
plt.ylabel("Winning Count")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = matches_df['venue'])
plt.title("Count of matches on each venue")
plt.xlabel("Venue Name")
plt.ylabel("Matches Count")
plt.xticks(rotation='vertical')
plt.show()
```

```python
pd.value_counts(matches_df['toss_decision'])
```

```python
matches_df.groupby(['season','toss_decision'])['toss_decision'].count()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = matches_df['season'], hue=matches_df['toss_decision'])
plt.title("Toss decision for each year")
plt.xlabel("Season")
plt.ylabel("Toss Decision Count")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.value_counts(matches_df['player_of_match']).head(10).plot.bar(width=0.8, color=sns.color_palette('inferno')
```

```python
deliveries_df.groupby(['match_id', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(width=0.8, color=sns.color_palette('inferno'))
```
