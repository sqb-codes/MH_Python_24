```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,20)
y = np.random.randint(10,100,19)
```

```python
plt.figure(figsize=(10,5))
# plt.plot(x,y)
plt.plot(x,y,'-x')
plt.title("My First Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
data = {
    "names" : ["John", "Max", "Sam", "Nick", "Adam", "Steve", "Alex"],
    "marks" : [78,88,86,45,90,37,55]
}

df = pd.DataFrame(data)
```

```python
plt.bar(df["names"], df["marks"], color='red')
plt.title("Marks of Students", color='blue')
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
plt.barh(df["names"], df["marks"], color='red')
plt.title("Marks of Students", color='blue')
plt.ylabel("Names")
plt.xlabel("Marks")
plt.show()
```

```python
parties = ['BJP', 'CONG', 'AAP', 'Others']
seats = [320, 110, 65, 82]
```

```python
plt.figure(figsize=(7,7))
plt.pie(seats, labels=parties, startangle=90, counterclock=False, autopct='%1.2f%%', shadow=True, explode=(0,0.2,0,0))
plt.title("Seats won by each party")
plt.legend(loc='upper right')
plt.show()
```
