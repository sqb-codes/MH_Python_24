```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('headbrain.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
X = np.reshape(X, (-1, 1))
y = np.reshape(y, (-1, 1))

minmax = MinMaxScaler()
X_scale = minmax.fit_transform(X)
```

```python
y_scale = minmax.fit_transform(y)
```

```python
X_scale[:5]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_scale, y_scale)
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```

```python
def gradientDescent(epochs, alpha):
    n = len(X)
    slope, inter = 0, 0
    for i in range(epochs):
        y_pred = X_scale * slope + inter
        loss = y_pred - y_scale
        gradSlope = (2/n) * np.dot(loss.T, X_scale)
        gradInter = (2/n) * np.sum(loss)
        slope = slope - gradSlope * alpha
        inter = inter - gradInter * alpha
        
    return slope, inter
```

```python
epochs = 10000
alpha = 0.01
slope, inter = gradientDescent(epochs, alpha)
```

```python
slope, inter
```

```python
n = len(X)
predictions = []
for i in range(n):
    # y = mx + c
    y_pred = slope[0][0] * X_scale[i] + inter
    predictions.append(y_pred)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_scale, y_scale)
plt.plot(X_scale, predictions, 'r')
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```
