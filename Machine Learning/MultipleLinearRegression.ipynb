```python
X = df[['Open', 'High', 'Low']].values
y = df['Close'].values
```

```python
# y = mx + c

# x = [X1, X2, X3..Xn]
# y = m1x1 + m2x2 + m3x3 + ... + mxn + c
```

```python
plt.boxplot(X)
plt.show()
```

```python
# x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
threshold = 0.20
n = len(X)
test_rows_n = int(n * threshold)
train_rows_n = n - test_rows_n
X_train = X[:train_rows_n]
y_train = y[:train_rows_n]
X_test = X[train_rows_n:]
y_test = y[train_rows_n:]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```
