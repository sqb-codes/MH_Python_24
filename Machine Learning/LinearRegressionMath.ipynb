```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('headbrain.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
mean_x, mean_y
```

```python
numer = 0
denom = 0

n = len(X)

for i in range(n):
    numer += (y[i] - mean_y) * (X[i] - mean_x)
    denom += (X[i] - mean_x) ** 2
    
slope = numer / denom
intercept = mean_y - slope * mean_x
```

```python
slope, intercept
```

```python
predictions = []
for i in range(n):
    # y = mx + c
    y_pred = slope * X[i] + intercept
    predictions.append(int(y_pred))
```

```python
predictions[:5]
```

```python
y[:5]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, predictions, 'r')
plt.xlabel("Head Size")
plt.ylabel("Brain Weight")
plt.show()
```

```python
# Mean Square Error
# MSE = sum((pred[i] - y[i]) ** 2)

error = 0
for i in range(n):
    error += (predictions[i] - y[i]) ** 2
    
mse = error / n
```

```python
print("MSE :",mse)
```

```python
#RMSE - Root MSE
np.sqrt(mse)
```
